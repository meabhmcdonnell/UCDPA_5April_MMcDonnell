```python
count2020=pd.read_csv("2020countdetails.csv", index_col=["Constituency Number"])
count2020.head()
```

```python
#merging candidate details with count details to get extra information such as total electorate, seats filled, total poll. Distinguishing repeated column names between 2 data sets with cand and count
candidate_count20 = data2020.merge(count2020,on="Constituency Number", suffixes=('_cand','count'))
candidate_count20.head()
```
